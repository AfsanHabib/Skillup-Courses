#### Matplotlib

**Matplotlib** is a comprehensive library for creating static, animated, and interactive visualizations in Python.
To know more about matplotlib library, refer to the link [here](https://matplotlib.org/)

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

Before plotting the ring chart for gdp data for countries in North America, we will see how a ring chart looks like for a  simple one dimensional data and plot it using the `pieplot()` function.

```python
# create data
dataset=[17,10,6,40]
 
# Create a pieplot
plt.pie(dataset)
 
# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()  #get current figure
p.gca().add_artist(my_circle) 
# The add_artist() is used to add an Artist to the axes, and return the artist.
# The artist is the object that knows how to take the Renderer and use it to put ink on the canvas. 
# Everything you see on a Matplotlib figure is an artist instance. 
plt.show()
```

This is what a basic ring plot looks like. 

The  trick is to make a pie chart and then add a white circle in the middle of the pie chart created. 
Another option is to use the `radius` and `width` arguments. 

### Add Annotations

```python
# create data
names = 'Category A', 'Category B', 'Category C', 'Category D' #create a list for labels
dataset=[17,10,6,40]
 
# Create a pieplot
plt.pie(dataset, labels=names) #use 'labels' argument
 
# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()  #get current figure
p.gca().add_artist(my_circle) 
# The add_artist() is used to add an Artist to the axes, and return the artist.
# The artist is the object that knows how to take the Renderer and use it to put ink on the canvas. 
# Everything you see on a Matplotlib figure is an artist instance. 
plt.show()
```

### Custom Color

```python
# create data
dataset=[17,10,6,40]
names = 'Category A', 'Category B', 'Category C', 'Category D' #create a list for labels
# Create a pieplot
plt.pie(dataset, colors=['red','green','blue','yellow'], labels=names) #explicitly specify colors using colors=[]
 
# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()  #get current figure
p.gca().add_artist(my_circle) 
# The add_artist() is used to add an Artist to the axes, and return the artist.
# The artist is the object that knows how to take the Renderer and use it to put ink on the canvas. 
# Everything you see on a Matplotlib figure is an artist instance. 
plt.show()
```

### Custom Wedges

```python
# create data
dataset=[17,10,6,40]
names = 'Category A', 'Category B', 'Category C', 'Category D' #create a list for labels
# Create a pieplot
plt.pie(dataset, wedgeprops= {'linewidth':5, 'edgecolor':'white'}, labels=names) #explicitly specify colors using colors=[]
 
# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()  #get current figure
p.gca().add_artist(my_circle) 
# The add_artist() is used to add an Artist to the axes, and return the artist.
# The artist is the object that knows how to take the Renderer and use it to put ink on the canvas. 
# Everything you see on a Matplotlib figure is an artist instance. 
plt.show()
```

#### panda Basics

We will now import the **pandas** module.

```python
import pandas as pd #we can also plot data directly from a Pandas framework
```

```python
df_namerica = pd.read_csv('https://raw.githubusercontent.com/arpgpt/data-visualization/master/north_america_pie.csv', index_col='country')
# df_namerica = [10,30,60]
```

We will use *pandas* `groupby` method to summarize the data by `Country`. The general process of `groupby` involves the following steps:

1. **Split:** Splitting the data into groups based on some criteria.
2. **Apply:** Applying a function to each group independently:
       .sum()
       .count()
       .mean() 
       .std() 
       .aggregate()
       .apply()
       .etc..
3. **Combine:** Combining the results into a data structure.

<img src = https://ibm.box.com/shared/static/tkfhxqkehfzpclco8f0eazhie33uxj9j.png,   height=400, align = "center">


```python
# group countries by continents and apply sum() function 
df_countries = df_namerica.groupby('country', axis=0).sum()

# note: the output of the groupby method is a `groupby' object. 
# we can not use it further until we apply a function (eg .sum())
print(type(df_namerica.groupby('country', axis=0)))

df_countries
```

We will now pass in `kind = 'pie'` keyword, along with the following additional parameters:
- `autopct` -  is a string or function used to label the wedges with their numeric value. The label will be placed inside the wedge. If it is a format string, the label will be `fmt%pct`.
- `startangle` - rotates the start of the pie chart by angle degrees counterclockwise from the x-axis.
- `shadow` - Draws a shadow beneath the pie (to give a 3D feel).

```python
# autopct create %, start angle represent starting point
df_countries['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90° (Africa)
                            shadow=True,       # add shadow      
                            )

plt.title('GDP of North America by Countries [1952 - 2007]')
plt.axis('equal') # Sets the pie chart to look like a circle.

my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```
